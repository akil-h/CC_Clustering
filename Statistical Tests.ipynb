##### The aim is to conduct statistical tests between each cluster

```python
# Add cluster labels to the data
cluster_labels = pd.DataFrame(Clustering.pca_cluster_labels)
data['Cluster_Labels'] = cluster_labels.replace({0: 'Cluster 1', 1: 'Cluster 2', 2: 'Cluster 3'})
data
```

```python
data.describe().T
```

```python
# Next: segment data into 3 dataframe subsets

Cluster1_data = data[data['Cluster_Labels'] == "Cluster 1"]
Cluster2_data = data[data['Cluster_Labels'] == "Cluster 2"]
Cluster3_data = data[data['Cluster_Labels'] == "Cluster 3"]
```

```python
print("There are {} observations in Cluster 1".format(len(Cluster1_data)))
print("There are {} observations in Cluster 2".format(len(Cluster2_data)))
print("There are {} observations in Cluster 3".format(len(Cluster3_data)))
```

```python
from scipy import stats

# perform an ANOVA test on feature means between all clusters on 
Credit_limit = 'CREDIT_LIMIT'
statistic, p_value = stats.f_oneway(Cluster1_data[Credit_limit], Cluster2_data[Credit_limit], Cluster3_data[Credit_limit])
print(f"F-value: {statistic}")
print(f"p-value: {p_value}")
```

##### The high F-value tells us that there exist statistically significant differences in the mean credit limits of cardholders between clusters. Let's investigate to see which clusters have the highest and lowest mean credit limits:

```python
print(f'Cluster 1: \n', Cluster1_data["CREDIT_LIMIT"].describe(), '\n')
print(f'Cluster 2: \n', Cluster2_data["CREDIT_LIMIT"].describe(), '\n')
print(f'Cluster 3: \n', Cluster3_data["CREDIT_LIMIT"].describe())
```

##### The mean cardholder credit limit is highest in Cluster 3 at $6737.96 compared to the lowest in Cluster 2 at $3950.54
